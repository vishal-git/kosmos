### Create image caption and identify entities within the image using the [Kosmos-2](https://huggingface.co/docs/transformers/main/model_doc/kosmos-2) model from HuggingFace.

Before running this Notebook, copy some images in the `images` directory. (Currently, there are three images included with this repo.)

# Setup

We will import two functions from `utils.py`:

1. `random_file()`: Randomly selects an image from the `images` directory.
2. `get_tags()`: Runs the `kosmos-2` model to get an image caption and image entities.

```python
import sys; sys.path.append('../src')
from utils import random_file, get_tags
```

#### Let's grab a random image from the `images` directory, display the image with the caption and tags/entities.

```python
caption, entities = get_tags(random_file('../images/*.jpg'), True)

print(f'Caption: {caption}')
print('Tags:')
for e in entities:
    print(e[0])
```

*Note that the prompt that's used here is: "<grounding> An image of".*

#### Now let's run this for ten random images.

```python
for i in range(10):
    caption, entities = get_tags(random_file('../images/*.jpg'), True)
    print(f'Caption: {caption}')
    print('Tags:')
    for e in entities:
        print(e[0])
```

Let's change the prompt and see if we can get more information about the image.

```python
caption, entities = get_tags('../images/2400207.jpg', True)
print(caption, entities)
```

```python
caption, entities = get_tags('../images/2400207.jpg', True, '<grounding>Describe this image in detail:')
print(caption, entities)
```
